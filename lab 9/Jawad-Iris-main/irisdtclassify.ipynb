```python
df = pd.read_csv('/kaggle/input/iris-dataset/iris.csv')
print(df.head())
print(df.info())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# Assuming df is your dataframe
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']  # sklearn handles string labels automatically

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Evaluate accuracy
accuracy = clf.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")

# Plot the decision tree
plt.figure(figsize=(20, 10))
tree.plot_tree(
    clf,
    feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    class_names=['setosa', 'versicolor', 'virginica'],
    filled=True,
    rounded=True
)
plt.savefig('decision_tree.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=clf.classes_)
plt.show()
```

```python
import joblib
joblib.dump(clf, "model.pkl")
loaded_model = joblib.load("model.pkl")

# Create DataFrame with proper feature names
input_df = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]],
                        columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

# Predict
prediction = loaded_model.predict(input_df)

print(prediction)

```
