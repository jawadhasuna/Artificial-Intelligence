```python
data=pd.read_csv('/kaggle/input/house-price-prediction-dataset/House Price Prediction Dataset.csv')
data.head()
data=data.drop_duplicates()
data=data[['Area','Bedrooms','Floors','Price']]
from sklearn.preprocessing import StandardScaler
sx = StandardScaler()
sy = StandardScaler()

X_scaled = sx.fit_transform(data[['Area','Bedrooms','Floors']])
y_scaled = sy.fit_transform(data[['Price']].values.reshape(-1,1))

xtrain,xtest,ytrain,ytest= train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
model=LinearRegression()
model.fit(xtrain,ytrain)
y_pred_scaled = model.predict(xtest)
# Inverse transform to get original prices
ypred = sy.inverse_transform(y_pred_scaled)
y_test_orig = sy.inverse_transform(ytest)
from sklearn.metrics import r2_score, mean_squared_error
print("R² Score:", r2_score(y_test_orig , ypred))
print("MSE:", mean_squared_error(y_test_orig , ypred))
```
