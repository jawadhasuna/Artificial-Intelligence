```python
df=pd.read_csv("/kaggle/input/ice-cream-sales-dataset/Ice Cream.csv")
print(df.columns)
x = df["Temperature"].values
y = df["Revenue"].values
df.head()
```

```python
m = len(y)
theta0 = 0.0
theta1 = 0.0
alpha = 0.001
epochs = 1000
```

```python
for epoch in range(epochs):
    y_pred = theta0 + theta1 * x
    error = y_pred - y

    grad0 = (1/m) * np.sum(error)
    grad1 = (1/m) * np.sum(error * x)

    theta0 = theta0 - alpha * grad0
    theta1 = theta1 - alpha * grad1

    if epoch % 100 == 0:
        cost = (1/(2*m)) * np.sum(error**2)
        print(f"Epoch {epoch:4d}: cost = {cost:.4f}, θ0 = {theta0:.4f}, θ1 = {theta1:.4f}")
```

```python
print("Final parameters:")
print("θ0 =", theta0)
print("θ1 =", theta1)
```
