```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Ice Cream.csv")
print(df.columns)
x = df["Temperature"].values
y = df["Revenue"].values
```

```python
df.head()
```

```python
m = len(y)
```

```python
theta0 = 0.0
theta1 = 0.0
```

```python
alpha = 0.001
epochs = 500
```

```python
for epoch in range(epochs):
    y_pred = theta0 + theta1 * x
    error = y_pred - y

    grad0 = (1/m) * np.sum(error)
    grad1 = (1/m) * np.sum(error * x)

    theta0 = theta0 - alpha * grad0
    theta1 = theta1 - alpha * grad1

    if epoch % 100 == 0:
        cost = (1/(2*m)) * np.sum(error**2)
        print(f"Epoch {epoch:4d}: cost = {cost:.4f}, θ0 = {theta0:.4f}, θ1 = {theta1:.4f}")
```
