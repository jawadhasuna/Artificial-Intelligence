```python
data=pd.read_csv("/content/hpdata.csv")
data=data[['Area','Bedrooms','Floors','Price']]
print(data.columns)
#preprocess
data=data.drop_duplicates()
data['Area']=(data['Area']/(data['Area'].mean()/10)).round().astype(int)
x1=data['Area'].values
x2=data['Bedrooms'].values
x3=data['Floors'].values
y=data['Price'].values
# initialize parameters
m = len(y)
theta0 = 0.0
theta1 = 0.0
theta2= 0.0
theta3= 0.0
alpha = 0.0000001
epochs = 10000
for epoch in range(epochs):
    y_pred = theta0 + theta1 * x1+ theta2 * x2+theta3 * x3
    error = y_pred - y

    grad0 = (1/m) * np.sum(error)
    grad1 = (1/m) * np.sum(error * x1)
    grad2 = (1/m) * np.sum(error * x2)
    grad3 = (1/m) * np.sum(error * x3)

    theta0 = theta0 - alpha * grad0
    theta1 = theta1 - alpha * grad1
    theta2 = theta2 - alpha * grad2
    theta3 = theta3 - alpha * grad3

    if epoch % 100 == 0:
        cost = (1/(2*m)) * np.sum(error**2)
        print(f"Epoch {epoch:4d}: cost = {cost:.4f}, θ0 = {theta0:.4f}, θ1 = {theta1:.4f}, θ2 = {theta2:.4f}, θ3 = {theta3:.4f}")

```

```python
# Example: predict price for a house
new_house = [13,5, 3]  # Area, Bedrooms, Floors
predicted_price = theta0 + theta1*new_house[0] + theta2*new_house[1] + theta3*new_house[2]
print("Predicted Price:", predicted_price)
```
