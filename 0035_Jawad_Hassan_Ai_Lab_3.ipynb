```python
#graph
#g={{v1,v2,v3,v4,v5},{e1,e2,e3,e4,e5}}
# ============================================================
# üß† BFS EXAMPLES IN PYTHON ‚Äî SIMPLEST AND MOST COMMON
# ============================================================
# Includes:
#   1. BFS on Graph (Shortest path)
#   2. BFS on Grid / Maze
#   3. BFS on 8-Puzzle problem
#   4. BFS Generic Template
#   5. BFS on Rubik's Cube (Skeleton demonstration)
# ============================================================

from collections import deque
from typing import List, Tuple, Dict, Optional

# ============================================================
# 1Ô∏è‚É£ BFS ON GRAPH (SHORTEST PATH)
# ============================================================

def bfs_graph(adj: Dict[str, List[str]], start: str, goal: str) -> Optional[List[str]]:
    """Breadth-First Search on adjacency list graph."""
    q = deque([start])
    parent = {start: None}

    while q:
        node = q.popleft()
        if node == goal:
            # reconstruct path
            path = []
            while node is not None:
                path.append(node)
                node = parent[node]
            return list(reversed(path))
        for nb in adj.get(node, []):
            if nb not in parent:
                parent[nb] = node
                q.append(nb)
    return None


# ============================================================
# 2Ô∏è‚É£ BFS ON GRID (PATH FINDING IN MAZE)
# ============================================================

def bfs_grid(start: Tuple[int, int], goal: Tuple[int, int], grid: List[List[int]]) -> Optional[List[Tuple[int, int]]]:
    """BFS in 2D grid. 0 = free cell, 1 = blocked."""
    R, C = len(grid), len(grid[0])
    q = deque([start])
    parent = {start: None}
    dirs = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    while q:
        r, c = q.popleft()
        if (r, c) == goal:
            path = []
            while (r, c) is not None:
                path.append((r, c))
                (r, c) = parent[(r, c)]
            return list(reversed(path))
        for dr, dc in dirs:
            nr, nc = r + dr, c + dc
            if 0 <= nr < R and 0 <= nc < C and grid[nr][nc] == 0 and (nr, nc) not in parent:
                parent[(nr, nc)] = (r, c)
                q.append((nr, nc))
    return None


# ============================================================
# 3Ô∏è‚É£ BFS ON 8-PUZZLE PROBLEM
# ============================================================

def bfs_8puzzle(start: Tuple[int, ...], goal: Tuple[int, ...]) -> Optional[List[Tuple[int, ...]]]:
    """8-Puzzle solver using BFS."""
    q = deque([start])
    parent = {start: None}

    while q:
        state = q.popleft()
        if state == goal:
            path = []
            while state is not None:
                path.append(state)
                state = parent[state]
            return list(reversed(path))

        # Find blank (0)
        z = state.index(0)
        r, c = divmod(z, 3)
        for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                new = list(state)
                nz = nr * 3 + nc
                new[z], new[nz] = new[nz], new[z]
                new_t = tuple(new)
                if new_t not in parent:
                    parent[new_t] = state
                    q.append(new_t)
    return None


# ============================================================
# 4Ô∏è‚É£ GENERIC BFS TEMPLATE
# ============================================================

def bfs_generic(start, is_goal_fn, neighbors_fn, max_nodes=100000):
    """Generic BFS ‚Äî use this for any problem."""
    q = deque([start])
    parent = {start: None}
    explored = 0

    while q:
        explored += 1
        if explored > max_nodes:
            return None, "‚ùå Node limit reached!"

        node = q.popleft()
        if is_goal_fn(node):
            # reconstruct path
            path = []
            while node is not None:
                path.append(node)
                node = parent[node]
            return list(reversed(path)), f"‚úÖ Explored {explored} nodes"

        for nb in neighbors_fn(node):
            if nb not in parent:
                parent[nb] = node
                q.append(nb)

    return None, f"‚ùå No path found after {explored} nodes"


# ============================================================
# 5Ô∏è‚É£ RUBIK‚ÄôS CUBE (SKELETON EXAMPLE)
# ============================================================
# ‚ö†Ô∏è Full Rubik's cube BFS is NOT feasible (4.3e19 states)
# But here‚Äôs a small skeleton for demonstration.

def rubik_neighbors_skeleton(state):
    """Dummy neighbor function ‚Äî for demo only."""
    a, b = state
    return [(a + 1, b), (a, b + 1), (a + 1, b + 1)]


# ============================================================
# üß© RUN EXAMPLES
# ============================================================

if __name__ == "__main__":
    print("\n==============================")
    print("1Ô∏è‚É£ GRAPH BFS EXAMPLE")
    print("==============================")

    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E'],
        'C': ['A', 'F'],
        'D': ['B'],
        'E': ['B', 'F'],
        'F': ['C', 'E', 'G'],
        'G': ['F']
    }
    print("Path A ‚Üí G:", bfs_graph(graph, 'A', 'G'))

    print("\n==============================")
    print("2Ô∏è‚É£ GRID BFS EXAMPLE")
    print("==============================")

    grid = [
        [0, 0, 0, 0],
        [1, 1, 0, 1],
        [0, 0, 0, 0],
        [0, 1, 1, 0]
    ]
    start, goal = (0, 0), (3, 3)
    print("Path from", start, "to", goal, ":", bfs_grid(start, goal, grid))

    print("\n==============================")
    print("3Ô∏è‚É£ 8-PUZZLE BFS EXAMPLE")
    print("==============================")

    start_8 = (1, 2, 3, 4, 5, 6, 7, 0, 8)
    goal_8 = (1, 2, 3, 4, 5, 6, 7, 8, 0)
    sol = bfs_8puzzle(start_8, goal_8)
    if sol:
        print("Steps to solve:", len(sol) - 1)
        for s in sol:
            print(s)
    else:
        print("‚ùå No solution found")

    print("\n==============================")
    print("4Ô∏è‚É£ GENERIC BFS DEMO (RUBIK‚ÄôS CUBE SKELETON)")
    print("==============================")

    start = (0, 0)
    goal_check = lambda s: s == (2, 2)
    path, status = bfs_generic(start, goal_check, rubik_neighbors_skeleton, max_nodes=100)
    print("Result path:", path)
    print("Status:", status)

    print("\n‚úÖ BFS Examples Completed Successfully ‚úÖ")
```

```python
class Node:
    def __init__(self, name, parent=None, actions=None, cost=0):
        self.name = name
        self.parent = parent
        self.actions = actions or []
        self.cost = cost
graph1={
    'A': Node('A', None, [('B', 6), ('C', 9), ('E', 1)],0),
    'B': Node('B', None, [('A', 6), ('D', 3), ('E', 4)], 0),
    'C': Node('C', None, [('A', 9), ('F', 2), ('G', 3)], 0),
    'D': Node('D', None, [('B', 3), ('E', 5), ('F', 7)], 0),
    'E': Node('E', None, [('A', 1), ('B', 4), ('D', 5), ('F', 6)], 0),
    'F': Node('F', None, [('C', 2), ('E', 6), ('D', 7)], 0),
    'G':('G',None,[('C',3)],0)}

from collections import deque

def bfs_with_cost(graph, start, goal):
    visited = []
    queue = deque([(start, 0)])  # (node_name, total_cost)

    while queue:
        node_name, cost = queue.popleft()
        if node_name not in visited:
            visited.append(node_name)
            print(f"Visited: {node_name}, Total Cost: {cost}")

            if node_name == goal:
                print(f"\n‚úÖ Reached Goal '{goal}' with Total Cost = {cost}")
                return cost

            # Enqueue all neighbours with updated costs
            for (neighbor, weight) in graph[node_name].actions:
                if neighbor not in visited:
                    queue.append((neighbor, cost + weight))

    print("\n‚ùå Goal not found.")
    return None
bfs_with_cost(graph1, 'A', 'G')
bfs_with_cost(graph1, 'A', 'F')
bfs_with_cost(graph1, 'E', 'G')
bfs_with_cost(graph1, 'B', 'F')
```

```python
# Python program to print DFS traversal for a complete graph
from collections import defaultdict

# This class represents a directed graph using adjacency list representation
class Graph:

    # Constructor
    def __init__(self):
        # Default dictionary to store graph
        self.graph = defaultdict(list)

    # Function to add an edge to the graph
    def addEdge(self, u, v):
        self.graph[u].append(v)

    # A function used by DFS
    def DFSUtil(self, v, visited):
        # Mark the current node as visited and print it
        visited[v] = True
        print(v, end=' ')

        # Recur for all the vertices adjacent to this vertex
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFSUtil(i, visited)

    # The function to do DFS traversal
    def DFS(self):
        V = len(self.graph)  # Total number of vertices
        # Mark all the vertices as not visited
        visited = [False] * V

        # Call the recursive helper function to print DFS traversal
        # starting from all vertices one by one
        for i in range(V):
            if visited[i] == False:
                self.DFSUtil(i, visited)


# Create a graph given in the example
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal:")
g.DFS()
```

```python
#task2
graph={
    'Arad':        [('Zerind',75),('Sibiu', 140), ('Timisoara', 118)],
    'Zerind':      [('Arad', 75), ('Oradea', 71)],
    'Oradea':      [('Zerind', 71), ('Sibiu', 151)],
    'Sibiu':       [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara':   [('Arad', 118), ('Lugoj', 111)],
    'Lugoj':       [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia':     [('Lugoj', 70), ('Dobreta', 75)],
    'Dobreta':     [('Mehadia', 75), ('Craiova', 120)],
    'Craiova':     [('Dobreta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras':     [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti':     [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest':   [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu':     [('Bucharest', 90)],
    'Urziceni':    [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Hirsova':     [('Urziceni', 98), ('Eforie', 86)],
    'Eforie':      [('Hirsova', 86)],
    'Vaslui':      [('Urziceni', 142), ('Iasi', 92)],
    'Iasi':        [('Vaslui', 92), ('Neamt', 87)],
    'Neamt':       [('Iasi', 87)]
}
from collections import deque

def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [start], 0)])  # (node, path, total_cost)

    while queue:
        node, path, cost = queue.popleft()

        if node == goal:
            return path, cost

        visited.add(node)

        for neighbor, distance in graph[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor], cost + distance))

    return None, float('inf')


# Run BFS
path, total_cost = bfs(graph, 'Arad', 'Bucharest')
print("BFS Path:", " ‚Üí ".join(path))
print("Total Distance:", total_cost)
#dfs
def dfs(graph, start, goal):
    stack = [(start, [start], 0)]  # (node, path, total_cost)
    visited = set()

    while stack:
        node, path, cost = stack.pop()

        if node == goal:
            return path, cost

        visited.add(node)

        for neighbor, distance in reversed(graph[node]):  # reversed for consistent path order
            if neighbor not in visited:
                stack.append((neighbor, path + [neighbor], cost + distance))

    return None, float('inf')


# Run DFS
path, total_cost = dfs(graph, 'Arad', 'Bucharest')
print("\nDFS Path:", " ‚Üí ".join(path))
print("Total Distance:", total_cost)
```
