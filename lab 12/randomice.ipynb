```python
# 1. Load Ice Cream dataset from Kaggle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Load dataset from Kaggle input
df = pd.read_csv("/kaggle/input/ice-cream-sales-dataset/Ice Cream.csv")

# Show first few rows to inspect
print(df.head())
print(df.columns)
# 2. Separate independent & dependent variables
# Assuming 'Temperature' as input and 'Sales' (or similar name) as target
X = df[['Temperature']]       # Independent variable
y = df['Revenue']               # Dependent variable

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
# 3. Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)

# 4 & 5. Random forest regressor setup
rf_regressor = RandomForestRegressor(
    n_estimators=10,
    random_state=0
)

# 6. Train model
rf_regressor.fit(X_train.values.reshape(-1, 1), y_train)
# 7. Predict
y_pred = rf_regressor.predict(X_test.values.reshape(-1, 1))

# Compare actual vs predicted
results = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})
print(results)
# 8. Plot results
plt.scatter(X, y, color='blue', label='Actual data')

plt.scatter(X, y, color='blue', label='Actual data')

# Fine grid for smooth curve
X_grid = np.arange(X.values.min(), X.values.max(), 0.1).reshape(-1, 1)
plt.plot(X_grid, rf_regressor.predict(X_grid), color='red', label='Prediction')

plt.title("Random Forest Regression - Ice Cream Sales")
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
