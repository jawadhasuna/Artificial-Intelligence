```python
#water jug problem
from collections import deque

# Function to check goal state
def is_goal(state):
    # Goal: exactly 2 liters in 4L jug
    return state[0] == 2 or state[1] == 2

# Function to get next possible states
def get_successors(state):
    a, b = state  # a = 4L jug, b = 3L jug
    successors = []

    # Capacities
    A, B = 4, 3

    # Possible actions
    # 1. Fill any jug
    successors.append((A, b))  # Fill 4L
    successors.append((a, B))  # Fill 3L

    # 2. Empty any jug
    successors.append((0, b))  # Empty 4L
    successors.append((a, 0))  # Empty 3L

    # 3. Pour 4L → 3L
    pour = min(a, B - b)
    successors.append((a - pour, b + pour))

    # 4. Pour 3L → 4L
    pour = min(b, A - a)
    successors.append((a + pour, b - pour))

    return successors


# ---------------- BFS Algorithm ----------------
def bfs():
    start = (0, 0)
    queue = deque([[start]])
    visited = set([start])

    while queue:
        path = queue.popleft()
        current = path[-1]

        if is_goal(current):
            print("BFS Solution Path:", path)
            return path

        for next_state in get_successors(current):
            if next_state not in visited:
                visited.add(next_state)
                queue.append(path + [next_state])

    print("No solution found using BFS.")
    return None


# ---------------- DFS Algorithm ----------------
def dfs():
    start = (0, 0)
    stack = [[start]]
    visited = set([start])

    while stack:
        path = stack.pop()
        current = path[-1]

        if is_goal(current):
            print("DFS Solution Path:", path)
            return path

        for next_state in get_successors(current):
            if next_state not in visited:
                visited.add(next_state)
                stack.append(path + [next_state])

    print("No solution found using DFS.")
    return None


# ---------------- Main Program ----------------
print("WATER JUG PROBLEM (4L, 3L)")

bfs_path = bfs()
print("\n----------------------------\n")
dfs_path = dfs()

```
