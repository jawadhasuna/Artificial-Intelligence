```python
#graph
#g={{V1,V2,V3,V4,V5},{E1,E2,E3,E4,E5}}
graph={
    '5':['3','7'],
    '3':['2','4'],
    '7':['8'],
     '2':[],
    '4':['8'],
    '8':[]
}
visited = []  # List for visited nodes.
queue = []    # Initialize a queue

def bfs(visited, graph, node):  # Function for BFS
    visited.append(node)
    queue.append(node)

    while queue:  # Creating loop to visit each node
        # Dequeue a vertex from queue and print it
        m = queue.pop(0)
        print(m, end=" ") # Added a space for readability

        # Get all adjacent vertices of the dequeued vertex m.
        # If an adjacent vertex has not been visited, then mark it visited and enqueue it.
        for neighbour in graph[m]: # Corrected syntax: 'for neighbour in graph[m]:'
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("Following is the Breadth-First Search traversal:")
bfs(visited, graph, '5') # function calling
print() # Add a newline after the output
```

```python
#task2
from collections import deque

CITY_GRAPH = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Mehadia': [('Lugoj', 70), ('Dobreta', 75)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Pitesti', 97), ('Craiova', 146)],
    'Dobreta': [('Mehadia', 75), ('Craiova', 120)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Craiova': [('Dobreta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Giurgiu': [('Bucharest', 90)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
    'Iasi': [('Vaslui', 92), ('Neamt', 87)],
    'Neamt': [('Iasi', 87)],
    'Eforie': [('Hirsova', 86)]
}

def reconstruct_path(parents, goal_node):
    path = []
    current = goal_node
    while current is not None:
        path.append(current)
        current = parents.get(current)
    return " -> ".join(path[::-1])

def bfs(graph, start, goal):
    queue = deque([start])
    parents = {start: None}
    visited = {start}

    print(f"\n--- Applying Breadth-First Search (BFS) from {start} to {goal} ---")

    while queue:
        current_node = queue.popleft()
        print(f"Expanding: {current_node}")

        if current_node == goal:
            print(f"\nGoal state '{goal}' reached!")
            path = reconstruct_path(parents, goal)
            return path, None

        neighbors = sorted([n[0] for n in graph.get(current_node, [])])

        for neighbor in neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                parents[neighbor] = current_node
                queue.append(neighbor)
                print(f"   -> Enqueued: {neighbor}")

    return "Path not found.", None


def dfs(graph, start, goal):
    stack = [(start, [start], 0)]
    visited = {start}

    print(f"\n--- Applying Depth-First Search (DFS) from {start} to {goal} ---")

    while stack:
        current_node, path, distance = stack.pop()
        print(f"Exploring: {current_node} (Distance: {distance})")

        if current_node == goal:
            print(f"\nGoal state '{goal}' reached!")
            return " -> ".join(path), distance

        neighbors = sorted(graph.get(current_node, []), key=lambda x: x[0], reverse=True)

        for neighbor, weight in neighbors:
            if neighbor not in visited:
                new_path = path + [neighbor]
                new_distance = distance + weight

                visited.add(neighbor)
                stack.append((neighbor, new_path, new_distance))
                print(f"   -> Pushed: {neighbor}")

    return "Path not found.", None

START_NODE = 'Arad'
GOAL_NODE = 'Bucharest'

print("--- Graph Search Algorithms for Roman Highway Map ---")
print(f"Start Node: {START_NODE}, Goal Node: {GOAL_NODE}\n")

bfs_path, _ = bfs(CITY_GRAPH, START_NODE, GOAL_NODE)

print("\n\n")
print(" Result for 2.ii: Breadth-First Search (BFS)")
print("")
print(f"Path (Shortest by edges): {bfs_path}")


dfs_path, total_distance = dfs(CITY_GRAPH, START_NODE, GOAL_NODE)

print("\n\n")
print(" Result for 3: Depth-First Search (DFS)")
print("")
print(f"Path Found: {dfs_path}")
print(f"Total Distance: {total_distance} km")
```
