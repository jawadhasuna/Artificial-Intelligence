```python
#task1
#part1: map
#map(grid) case
path=[
    ['#','#','.','.','G'],
    ['#','#','.','.','.'],
    ['.','#','#','.','.'],
    ['.','#','.','.','.'],
    ['.','S','.','.','.'],
    ]
# Display map
print("path of case where # is blocked and . is available:")
for i in path:
  print(i)
#part2:
#find bfs to go S TO G:
from collections import deque

# Find coordinates of S and G
for i in range(len(path)):
    for j in range(len(path[0])):
        if path[i][j] == 'S':
            start = (i, j)
        if path[i][j] == 'G':
            goal = (i, j)

# Directions: up, down, left, right
directions = [(-1,0),(1,0),(0,-1),(0,1)]
# BFS
def bfs(start, goal):
    queue = deque([(start, [start])])  # store (node, path_so_far)
    visited = set([start])

    while queue:
        (x, y), path_so_far = queue.popleft()

        if (x, y) == goal:
            return path_so_far

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if (0 <= nx < len(path) and 0 <= ny < len(path[0])
                and path[nx][ny] in ('.', 'G') and (nx, ny) not in visited):
                visited.add((nx, ny))
                queue.append(((nx, ny), path_so_far + [(nx, ny)]))
    return None


# DFS
def dfs(start, goal):
    stack = [(start, [start])]
    visited = set([start])

    while stack:
        (x, y), path_so_far = stack.pop()

        if (x, y) == goal:
            return path_so_far

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if (0 <= nx < len(path) and 0 <= ny < len(path[0])
                and path[nx][ny] in ('.', 'G') and (nx, ny) not in visited):
                visited.add((nx, ny))
                stack.append(((nx, ny), path_so_far + [(nx, ny)]))
    return None

# Run both searches
bfs_path = bfs(start, goal)
dfs_path = dfs(start, goal)

print("\nBFS Path:")
print(bfs_path if bfs_path else "No path found")

print("\nDFS Path :")
print(dfs_path if dfs_path else "No path found")
```

```python
#part4
#genetic algo:
import random

# Map/Grid
path_grid = [
    ['#','#','#','.','G'],
    ['#','#','#','.','#'],
    ['#','#','#','.','#'],
    ['#','#','.','.','#'],
    ['#','S','.','#','#']
]

rows, cols = len(path_grid), len(path_grid[0])

# Directions
moves = ['U', 'D', 'L', 'R']
dir_map = {'U': (-1,0), 'D': (1,0), 'L': (0,-1), 'R': (0,1)}

# Find Start and Goal
for i in range(rows):
    for j in range(cols):
        if path_grid[i][j] == 'S':
            start = (i,j)
        if path_grid[i][j] == 'G':
            goal = (i,j)

# Parameters
POP_SIZE = 50
CHROMOSOME_LEN = 20  # max moves in path
GENERATIONS = 1000
MUTATION_RATE = 0.2

# Generate initial population
def generate_population():
    return [[random.choice(moves) for _ in range(CHROMOSOME_LEN)] for _ in range(POP_SIZE)]

# Fitness: shorter distance to goal is better, big penalty if hits wall
def fitness(chromosome):
    x, y = start
    penalty = 0
    for move in chromosome:
        dx, dy = dir_map[move]
        nx, ny = x + dx, y + dy
        if 0 <= nx < rows and 0 <= ny < cols and path_grid[nx][ny] != '#':
            x, y = nx, ny
        else:
            penalty += 1  # hit wall
    # Manhattan distance to goal
    dist = abs(x - goal[0]) + abs(y - goal[1])
    return -dist - penalty  # higher is better

# Selection: top 50%
def selection(pop, scores):
    sorted_pop = [x for _, x in sorted(zip(scores, pop), reverse=True)]
    return sorted_pop[:POP_SIZE//2]

# Crossover: single point
def crossover(parent1, parent2):
    point = random.randint(1, CHROMOSOME_LEN-1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

# Mutation: random move
def mutate(chromosome):
    if random.random() < MUTATION_RATE:
        idx = random.randint(0, CHROMOSOME_LEN-1)
        chromosome[idx] = random.choice(moves)
    return chromosome

# Run GA
population = generate_population()

for gen in range(GENERATIONS):
    scores = [fitness(ch) for ch in population]

    # Check if goal reached
    if max(scores) == -0:  # 0 distance to goal, no wall penalty
        best_index = scores.index(max(scores))
        print(f"Goal reached in generation {gen}!")
        print("Path:", population[best_index])
        break

    # Selection
    selected = selection(population, scores)

    # Crossover & Mutation to create new population
    new_pop = []
    while len(new_pop) < POP_SIZE:
        p1, p2 = random.sample(selected, 2)
        c1, c2 = crossover(p1, p2)
        new_pop.append(mutate(c1))
        if len(new_pop) < POP_SIZE:
            new_pop.append(mutate(c2))

    population = new_pop
else:
    best_index = scores.index(max(scores))
    print("Goal not reached. Best path:")
    print("Path:", population[best_index])
```
