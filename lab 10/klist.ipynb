```python
 import numpy as np
 import matplotlib.pyplot as plt
 from numpy.random import random
 %matplotlib inline
 N = 1500
 mean1 = [6, 14]
 mean2 = [10, 6]
 mean3 = [14, 14]
 cov = [[3.5, 0], [0, 3.5]] # diagonal covariance
 np.random.seed(50)
 X = np.random.multivariate_normal(mean1, cov, int(N/6))
 X = np.concatenate((X, np.random.multivariate_normal(mean2, cov,
 int(N/6))))
 X = np.concatenate((X, np.random.multivariate_normal(mean3, cov,
 int(N/6))))
 X = np.concatenate((X, 20*np.random.rand(int(N/2),2)))
 Y = np.concatenate((np.ones(int(N/2)),np.zeros(int(N/2))))
 plt.plot(X[:int(N/2),0],X[:int(N/2),1],'r+',X[int(N/2):,0],X[int(
 N/2):,1],'k.',ms=4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Assuming X and Y are your dataset
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# List of neighbors to test
numNeighbors = [1, 5, 10, 15, 20, 25, 30]

# Store accuracies
train_acc = []
test_acc = []

# Loop over different k values
for k in numNeighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Predict train and test
    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)
    
    # Compute accuracies
    train_acc.append(accuracy_score(y_train, y_train_pred))
    test_acc.append(accuracy_score(y_test, y_test_pred))

# Plotting
plt.figure(figsize=(8,5))
plt.plot(numNeighbors, train_acc, 'r-o', label='Training Accuracy')
plt.plot(numNeighbors, test_acc, 'b--v', label='Testing Accuracy')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("KNN: Training vs Testing Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Pick the best k (for example, k=10 from the accuracy curve)
best_k = 10

# Train KNN
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)

# Create meshgrid for plotting decision boundary
h = 0.2  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict for each point in meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
cmap_light = ListedColormap(['#FFAAAA','#AAAAFF'])
cmap_bold = ListedColormap(['red','green'])

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap_light)

# Plot training points
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=cmap_bold, edgecolor='k', s=60, marker='o', label='Train')
# Plot testing points
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, cmap=cmap_bold, edgecolor='k', s=60, marker='^', label='Test')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title(f"KNN Decision Boundary (k={best_k})")
plt.legend()
plt.show()
```
