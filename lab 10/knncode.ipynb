```python
import math
from collections import Counter
from sklearn.datasets import load_iris
import numpy as np

# ----------------------------
# Euclidean Distance
# ----------------------------
def euclidean(a, b):
    return math.sqrt(sum((x-y)**2 for x, y in zip(a, b)))

# ----------------------------
# Normalize
# ----------------------------
def normalize(X):
    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    return (X - mins) / (maxs - mins), mins, maxs

# ----------------------------
# KNN Classification ONLY
# ----------------------------
def knn_classify(X, y, x_query, k):
    distances = [(euclidean(x, x_query), idx) for idx, x in enumerate(X)]
    distances.sort(key=lambda x: x[0])
    neighbors = distances[:k]
    labels = [y[idx] for (_, idx) in neighbors]

    # mode (most common)
    counts = Counter(labels)
    return counts.most_common(1)[0][0]   # guaranteed integer


# ----------------------------
# Main Program
# ----------------------------
print("=== KNN on IRIS DATASET ===")

# load iris
iris = load_iris()
X = iris.data
y = iris.target

# normalize
X_norm, mins, maxs = normalize(X)

# choose a query sample (first one)
x_query = X_norm[0]

# ask only for k
k = int(input("Enter value of k: "))

# predict
pred = knn_classify(X_norm, y, x_query, k)

print("\nPrediction for sample #0 (Iris):", pred)
print("Predicted class name:", iris.target_names[int(pred)])
```
